### Q1. Loan Approval: Classification or Regression?

*   **It's Classification:** Think of this as putting things into buckets. Since the bank is choosing between two distinct categories ('Approved' or 'Rejected'), it's a classification problem.
*   **Best Algorithm:** **Decision Tree** is the winner here. It works like a checklist (e.g., 'Is credit score > 700?'), making it very easy for bankers to explain why a loan was accepted or denied.

### Q2. Features vs. Labels (The 'Input' vs. 'Result')

*   **Features (The Inputs):** These are the clues or information you look at.
*   **Label (The Result):** This is the actual answer you are trying to guess.

**Example:** If you're trying to guess if a Netflix user will stay subscribed:
*   **Features:** What shows they watch, how many hours they spend on the app, and if they use a phone or TV.
*   **Label:** Did they stay? (Yes or No).

### Q3. How a Decision Tree Thinks

Imagine playing '20 Questions'. The model asks a series of simple 'Yes/No' questions to narrow things down:
1. It starts at the top (e.g., 'Does the customer have a discount code?').
2. Depending on the answer, it moves to the next question.
3. It keeps going until it reaches a final decision (e.g., 'They will likely buy this product').

### Q4. KNN: Learning from Neighbors

*   **Why Supervised?** KNN is like a student who learns by looking at a map of points that are already labeled. If you don't tell the model what the existing points are, it has no way of knowing what the new point should be.
*   **No Labels?** No, KNN can't classify without labels. Without labels, it wouldn't know if a neighbor is 'A' or 'B'—it would just see dots on a map.

### Q5. A Quick KNN Count

*   **The Scene:** You look at the 5 closest people to a data point. Three say 'Yes' and two say 'No'.
*   **The Prediction:** **Yes**.
*   **The Reason:** It's a simple majority vote. Since 'Yes' has more votes (3 vs 2), the model goes with 'Yes'.

### Q6. Explaining $y = mx + b$

*   **$y$:** What you want to find out (e.g., predicted house price).
*   **$x$:** The information you have (e.g., square footage).
*   **$m$:** The multiplier. It shows how much $y$ changes for every bit of $x$ (the 'steepness').
*   **$b$:** The starting point. It’s what $y$ is when $x$ is zero.

### Q7. Understanding Correlation

Correlation tells you if two things move together.

*   **+1 (Perfect Match):** When one goes up, the other goes up exactly the same way.
*   **0 (Random):** There is no connection. One moving tells you nothing about the other.
*   **-1 (Opposites):** When one goes up, the other goes down.

### Q8. The Danger of a 'False Negative'

*   **What happened?** The model made a mistake called a **False Negative**. It told a sick person they were healthy.
*   **Why it's scary:** In medicine, this is the worst kind of mistake. A person might leave the hospital without treatment because the computer said they were fine, even though they are actually sick.

### Q9. Confusion vs. Correlation

*   **Confusion Matrix:** A report card for a model. It shows exactly how many times the model got it right and exactly what kind of mistakes (like the one above) it made.
*   **Correlation Matrix:** A map showing how much different features (like age and height) relate to one another.

### Q10. The 'Big Idea' for Each Algorithm

*   **Decision Tree:** A logical flowchart of 'Yes/No' questions.
*   **KNN:** 'Show me who your neighbors are, and I’ll tell you who you are.'
*   **SVM:** Drawing the widest possible line to separate two groups.
*   **Linear Regression:** Drawing a single straight line through data points to find a trend.
*   **Neural Network:** A complex web of layers that learns patterns, inspired by how our brains work.
